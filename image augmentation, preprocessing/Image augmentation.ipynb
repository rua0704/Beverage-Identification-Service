```python
import cv2
import os
import numpy as np

path = "./224_Validation"
file_list = os.listdir(path)
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

def imwrite2(filename, img, params=None):
    try:
        ext = os.path.splitext(filename)[1] 
        result, n = cv2.imencode(ext, img, params) 
        if result: 
            with open(filename, mode='w+b') as f:
                n.tofile(f) 
            return True
        else: 
            return False
    except Exception as e:
        print(e) 
        return False

```

```python
for i in file_list:
    file_list2 = os.listdir(path+'/'+i)
    print(i)
    for k,j in enumerate(file_list2):
        img = cv2.imread(path+'/'+i+'/'+j)
        img90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) # 시계방향으로 90도 회전
        img180 = cv2.rotate(img, cv2.ROTATE_180) # 180도 회전
        img270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

        cv2.imwrite(r'./img100_rot/'+i+'/valid'+str(k)+'.jpg',img)
        cv2.imwrite(r'./img100_rot/'+i+'/90_test'+str(k)+'.jpg',img90)
        cv2.imwrite(r'./img100_rot/'+i+'/180_test'+str(k)+'.jpg',img180)
        cv2.imwrite(r'./img100_rot/'+i+'/270_test'+str(k)+'.jpg',img270)
```
